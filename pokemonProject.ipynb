```python
classes = pd.read_csv('combats.csv')
classes
```

```python
classes = classes.groupby('First_pokemon').head(10).reset_index(drop=True)
classes
```

```python
#add extra column to test data for ground, 1 if first pokemon wins, 0 if second pokemon wins
ground_truth = []
for row in classes.itertuples():
    if(row.First_pokemon == row.Winner):
        ground_truth.append(1)
    else:
        ground_truth.append(0)
        
classes['ground'] = ground_truth
classes
```

```python
arr = []
#poke_count = dict()
#grab id of first and second pokemon in combats.csv and compare the stats
for row in classes.itertuples():
    #reset new array after each iteration
    new = []
    firstPoke = row.First_pokemon
    secondPoke = row.Second_pokemon
   # poke_count[firstPoke] = poke_count.get(firstPoke, 0) + 1
    
    new.append(int(data[data.id == firstPoke].HP) - int(data[data.id == secondPoke].HP))
    new.append(int(data[data.id == firstPoke].Attack) - int(data[data.id == secondPoke].Attack))
    new.append(int(data[data.id == firstPoke].Defense) - int(data[data.id == secondPoke].Defense))
    new.append(int(data[data.id == firstPoke].SpAtk) - int(data[data.id == secondPoke].SpAtk))
    new.append(int(data[data.id == firstPoke].SpDef) - int(data[data.id == secondPoke].SpDef))
    new.append(int(data[data.id == firstPoke].Speed) - int(data[data.id == secondPoke].Speed))
    cumulative = sum(new)
    new.append(cumulative)
    arr.append(new)
arr
#for i in poke_count:
  #  print (i, poke_count[i])
```
